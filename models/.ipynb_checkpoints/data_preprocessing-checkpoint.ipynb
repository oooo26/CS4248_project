```python
import pandas as pd
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
import seaborn as sns
import re
import unidecode
import string
```

```python
train = pd.read_csv('fulltrain.csv',names = ['label', 'text'])
train
```

```python
fig, ax = plt.subplots(figsize=(4,4))
plt.pie(train['label'].value_counts(), labels=train['label'].value_counts().index, autopct='%1.2f%%', startangle=90, wedgeprops={'edgecolor' :'black'})
plt.title('Pie chart of classes')
plt.legend(loc='best')
plt.axis('equal')
plt.show()
```

May need to try data balancing algorithms to make the label more balance, SMOTE may help or there are some other data augmentation algorithms specifically used for natural language processing.

```python
# remove punctuation
train['text'] = train['text'].apply(lambda x: re.sub('[%s]' % re.escape(string.punctuation), '' , x))
train
```

```python
def remove_accented_chars(text):
    """remove accented characters from text, e.g. café"""
    text = unidecode.unidecode(text)
    return text

train['text'] = train['text'].apply(lambda x:remove_accented_chars(x))
train
```

```python
# lower case
def lower_case(sentence):
    sentence = str(sentence)
    sentence = sentence.lower()
    return sentence

train['text'] = train['text'].apply(lambda x:lower_case(x))
train
```

```python
def remove_html_tag(text):
    tag_regex = r"<[^>]+>"
    tag_pattern = re.compile(tag_regex)

    return tag_pattern.sub("", text)

train['text'] = train['text'].apply(lambda x: remove_html_tag(x))
train
```

```python
# remove stop words(This is not necessary, need to make sure removing stopwords will not cause bad influence to the classification before using it)

from nltk.corpus import stopwords
stop_words = set(stopwords.words('english'))

def remove_stopwords(text):
    return " ".join([word for word in str(text).split() if word not in stop_words])

train['text'] = train['text'].apply(lambda x: remove_stopwords(x))
train
```
